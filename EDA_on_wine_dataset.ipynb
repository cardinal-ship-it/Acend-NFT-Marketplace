```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Quality_wine = pd.read_csv("winequality-red.csv")
Quality_wine.head()
```

```python
Quality_wine.info()
```

# New Section

```python
Quality_wine.describe()
```

```python
Quality_wine.hist(bins = 10, figsize = (16,12))
plt.show()
```

```python
# Create pivot table for Quality_wine
columns = list(Quality_wine.columns).remove('quality')
Quality_wine.pivot_table(columns,['quality'], aggfunc=np.median)
```

```python
# check for correlation
Quality_wine.corr()
```

```python
# plot correlation matrix using heatmap to visualize
plt.figure(figsize=(12,16))
sns.heatmap(Quality_wine.corr(), cmap='bwr', annot=True) # annot True to display the correlation value in the graph
```

```python
# we visualize using countplot
plt.figure(figsize=(10,6))
sns.countplot(Quality_wine.quality);
```
